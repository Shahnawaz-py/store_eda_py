```python
# from google.colab import files
# upload = files.upload()

# upload.keys()

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('Sample - Superstore (1).csv', encoding = 'latin1')

# Basic Analysis

# Q1 Check first five columns, no.of rows and cols, data type, null values, duplicate values

df.head()

df.describe()

df.columns.value_counts().sum()

df.shape

df.dtypes

df.isnull().sum()

df.duplicated().sum()
```

```python
# Intermediate Analysis

# Q2 Total sales
df['Sales'].sum()

# Q3 Total profit
df['Profit'].sum()

# Q3 Average discount
df['Discount'].mean()

# Q4 Total orders
df['Order ID'].nunique()

# Q5 Number of unique cities, customers, categories
df['City'].nunique()

df['Customer Name'].nunique()

df['Category'].nunique()
```

```python
# Advance Analysis

# Q6 Total sales per category
df.groupby('Category')['Sales'].sum()

# Q7 Total profit per category
df.groupby('Category')['Profit'].sum()

# Q8 Top 10 products by sales
df.groupby('Product Name')['Sales'].sum().sort_values(ascending = False).head(10)

# Q9 Top 10 products by profit
df.groupby('Product Name')['Profit'].sum().sort_values(ascending = False).head(10)

# Q10 Top 10 cities with highest sales
df.groupby('City')['Sales'].sum().sort_values(ascending = False).head(10)

# Q11 Top 10 cities with lowest sales
df.groupby('City')['Sales'].sum().sort_values(ascending = True).head(10)

# Q12 Top 10 region by sales
df.groupby('Region')['Sales'].sum().sort_values(ascending = False).head(10)

# Q13 Top 10 region by profit
df.groupby('Region')['Profit'].sum().sort_values(ascending = False).head(10)

# Q14 Top ship mode
df['Ship Mode'].value_counts().head(1)

# Q15 Products with highest discount
df[df['Discount'] == df['Discount'].max()]
```

```python
# Data visualization

# Q1 Total sales per category

sales_per_cat = df.groupby('Category')['Sales'].sum()
plt.figure()
sales_per_cat.plot(kind = 'bar')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Total Sales per Category')
plt.show()
```

```python
# Q2 Total profit per category

sales_per_cat = df.groupby('Category')['Profit'].sum()
plt.figure()
sales_per_cat.plot(kind = 'bar')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.title('Total Profit per Category')
plt.show()
```

```python
# Q3 Top 10 products per sales

top10_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending = False).head(10)
plt.figure()
top10_products.plot(kind = 'bar')
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.title('Top 10 selling products')
plt.show()
```

```python
#Q4 Top 10 products with highest profits

plt.scatter(df['Sales'], df['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales vs Profit')
plt.show()
```

```python
#Q5 Top 10 cities with highest sales

top10_cities = df.groupby('City')['Sales'].sum().sort_values(ascending = False).head(10).reset_index()
plt.figure()
sns.barplot(data = top10_cities, x = 'City', y = 'Sales')
plt.xlabel('City')
plt.ylabel('Sales')
plt.title('City vs Sales')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Q6 Sales per region

sales_reg = df.groupby('Region')['Sales'].sum()
sales_reg.plot(kind = 'pie', autopct = '%1.1f%%')
plt.ylabel('')
plt.title('Sales vs Region')
plt.show()
```
